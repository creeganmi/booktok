```python
import requests
import os
import json
import pandas as pd
import time
import datetime
import dateutil
import numpy as np
from dateutil.relativedelta import relativedelta
from datetime import date, timedelta
from pprint import pprint
```

```python
# Inputs:
## List name encoded (from list of NYT lists)
## months to retrieve (here 3 months to today)
## date frequency (here 5 days)
```

```python
list_name = 'combined-print-and-e-book-fiction'

end = datetime.date.today()
start = end - relativedelta(months=3)
dates_in_range = [x.split(' ') for x in pd.date_range(start, end, freq='5D').strftime("%Y %m %d").tolist()]


def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
    ## reference: https://towardsdatascience.com/flattening-json-objects-in-python-f5343c794b10 

    
data = {'list_name':[],'bestsellers_date':[],'rank':[],'rank_last_week':[],
         'weeks_on_list':[],'title':[],'author':[], 'ISBN-13':[]}


for date in dates_in_range:
    query_url = 'https://api.nytimes.com/svc/books/v3/lists/'+str(date[0])+'-'+str(date[1])+'-'+str(date[2])+'/'+list_name+'.json?api-key=YOUR_API_KEY'
    overview = requests.get(query_url).json()
    time.sleep(6)
    
    results = overview['results']
    books = results['books']
    books_flat = flatten_json(books)
    
    for item in books:
        data['list_name'].append(results['list_name'])
        data['bestsellers_date'].append(results['bestsellers_date'])
        data['rank'].append('')
        data['rank_last_week'].append('')
        data['weeks_on_list'].append('')
        data['title'].append('')
        data['author'].append('')
        data['ISBN-13'].append('')

data = pd.DataFrame(data)
data
```

```python
for date in dates_in_range:
    for n in range(15):
        string_rank = str(n)+'_rank'
        data.loc[n,'rank'] = books_flat[string_rank]

        string_rank_last_week = str(n)+'_rank_last_week'
        data.loc[n,'rank_last_week'] = books_flat[string_rank_last_week]

        string_weeks = str(n)+'_weeks_on_list'
        data.loc[n,'weeks_on_list'] = books_flat[string_weeks]

        string_title = str(n)+'_title'
        data.loc[n,'title'] = books_flat[string_title]

        string_author = str(n)+'_author'
        data.loc[n,'author'] = books_flat[string_author]

        string_ISBN = str(n)+'_primary_isbn13'
        data.loc[n,'ISBN-13'] = books_flat[string_ISBN]

    
data
```
